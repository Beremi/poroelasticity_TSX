```python
import numpy as np
import matplotlib.pyplot as plt
from model_TOM import WrapperTOM
import h5py
wrapper = WrapperTOM()
means = np.array([24.81761039916851,
                  -25.588243558620302,
                  -25.588243558620302,
                  -25.588243558620302,
                  -25.588243558620302,
                  -41.95735729765881,
                  -28.14184673969454,
                  -28.14184673969454,
                  -35.04960201867667,
                  17.62217304773459,
                  16.213405830762643])
stds = np.array([2, 3, 3, 3, 4, 4, 4, 4, 4, 2, 2], dtype=float)
```

```python
sampler = 0
sample_num = 0

perturb = np.random.randn(11)
param = means + stds * perturb
data = wrapper.run(param)
# Create the HDF5 file
file_path = f'test_runs/sampler_{sampler}_sample_{sample_num}.h5'
with h5py.File(file_path, 'w') as file:
    # Save the 'param' variable
    file.create_dataset('param', data=param)
    
    # Save the 'data' variable
    file.create_dataset('data', data=data)
```
